```python
# Step 1: Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Step 2: Group by Date and sum Total Amount
daily_sales = df.groupby('Date')['Total Amount'].sum()

# Step 3: Show first 10 results
print(daily_sales.head(10))

# Step 4: Plot (we will add graphs later)
import matplotlib.pyplot as plt

daily_sales.plot(kind='line', figsize=(10,5), title="Daily Sales Trend")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.show()

```
